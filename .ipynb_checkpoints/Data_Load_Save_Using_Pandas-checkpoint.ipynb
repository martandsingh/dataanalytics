```python
#import pandas for reading csv files or any other datatype
import pandas as pd
```

```python
# You can check by printing panda object, whether pandas is loaded properly or not.
pd
```

```python
# Load data from csv. In this example we will load data from online CSV file. 
# You can also use for the local system, file
file_path = 'https://raw.githubusercontent.com/martandsingh/datasets/master/cars/imports-85.data' # Online CSV Path
#If you will open the above url, you will see there is no header column in the data.
# So we will set header as None using 'header' parameter.
df = pd.read_csv(file_path, header = None)
```

```python
# Lets check whether data is loaded or not
df
```

```python
# Above we can see as  header was None, We do not have any header.
# Numeric headers are very hard to deal with. So lets add some Proper labels
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
# So we have an list header, which contains columns names.
print(headers)
```

```python
# Now Lets Set headers
df.columns = headers
```

```python
# Print dataframe. You can see, we have proper header names.
df
```

```python
# Every time we print our dataframe, it gives you a huge output. 
# We can deal with this situation by simply printing head or tail.
df.head() # By default it prints top 5 rows
```

```python
df.head(10) # You can choose any number
```
