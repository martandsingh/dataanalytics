Data Wrangling is the process of converting data from the initial format to a format that may be better for analysis.

We all know data can come from any kind of source e.g. forms, text files, database, audio, video. Different source means different format. The main differene between data and information is data is raw and you process it to get useful information.

Raw or non-formatted data is very difficult to analyze and it affects the accuracy of your analytical model.
To overcome all these things we will format our data.

e.g. Suppose we have data for the most popular city in India, We have different sources. 
Source1 : New Delhi
Source2 : NewDelhi
source3: ND
source4: N.D

So we can see different sourc providing different answers for the same place New Delhi. Currently our data is non formatted, which is very difficult to analyze, predict and accurate.
<br/><i>
The complete process of purifying data for the further processing is called data preprocessing. it is sometimes called data cleaning and data wrangling. 
    </i>

<h2>Topics Covered:</h2>
<h3>1. Identifying & Handling Missing Values </h3> - Missing value can cause inaccuracy to your statistical model. So we will learn how to identify null values and resolve it.
<h3>2. Data Formatting</h3> - We need to format data in common format. Above popular city example.
<h3>3. Data Normalization </h3>- Data should be in same scale otherwise it will be very difficult to calculate.
e.g.
    Age         Salary
    20          3000000
    25          5000000
    30          100000
In above example we can see Age & Salary has a huge differenc in value. If we plot this on graph it will be very difficult to observe both values together. So we need to normalize both varibles with the same algorithm. So that there values drop to the same scale.    
<h3>4. Data Binning </h3> - We will divide data into small grous or bins. It can cateogries our data.
e.g. 
   Item          Price
   A             3000
   B             3500
   C             4000
   D             80000
 
 We can make bins like:
 Low Price : 3000-4000
 Medium Price: 4001-8000
 High Price: >8001
<h3>5. Turning Categorical Variable into Int</h3> - Categorical variable as the name suggest are the variables which can catgorized or can have fix set of values.
e.g. Gender: Male Or Female
Subject:: Eng, Math, History
In a computer USB Version : 1.0, 2.0, 3.0

but your computer only understand numerical value so in case of Gender it doesnt know that it is a categorical variable. So first we need to convert it into categorical value.

Above whole process is called <b>DATA PROCESSING </b> or <b>DATACLEANING</b>

```python
# imports
import pandas as pd # for data load
import matplotlib.pylab as plt # Grpahical represntation
import numpy as np # for data related operations
```

```python
# CSV file path. IT can be any resource.
file_name = 'https://raw.githubusercontent.com/martandsingh/datasets/master/cars/imports-85.data'
```

```python
# for the header and read_csv method refer our module 1.
# https://github.com/martandsingh/dataanalytics/blob/master/Data_Load_Save_Using_Pandas.ipynb
df = pd.read_csv(file_name, header=None)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head()
```

```python
# AS we can see there are few question marks, These are missing value. It can be caused by the algorithm
# Which generates these files or human error.
# Now we need to get rid of these missing values
# Step 1: replace ? or any other junk values to the pandas nan value.
# For this we will use pandas replace(old_value, new_value, inplace = True) method.
# inplace = True means we want to change our main data frame.
# df.replace('?', np.nan, inplace=True) is equals to the df = df.replace('?', np.nan)
df.replace('?', np.nan, inplace=True)
```

```python
# Now we can see instead of ?, there are nan values.
df.head()
```

# Congrats you have detected your missing values... :)

```python
# Lets apply. True means there is null values.
df.isnull().head()
```

```python
# In this case we are checking for not nullability so false means null values.
df.notnull().head()
```

```python
# Check the number of missing values in columns
df.isnull().sum()
```

1. Remove the rows containing missing values - It is good when most of the rows have missing values(so not our case).
2. Replace missing values with:
    a. Average/Mean
    b. Mode
    c.Median

```python
# Calculate mean, median, mode or standard deviation
# It is very easy to calculate statstical values in python pandas
# astype is to change the data type of the columns
df['price'].dtype
```

```python
# AS you can see the datatype of price column is object, we need to change it float as price can not be in strings
# So we will first change it to float using astype() then we will calculate the mean
# axis=0, 1. if you choose axis=0 that mean you are performing operation along rows,and 1 is for the column
df['price'].astype("float").mean(axis=0)
```

```python
df['price'].astype("float").median(axis=0) # axis=0 means median or rows
```

```python
df['price'].astype("float").std()
```

```python
# Now as we know how to calculate mean, mode, median and standard deviation(std). Lets resolve missing values.
# First we will resolve normalized-losses 
avg_of_normalized_loss = df['normalized-losses'].astype('float').mean()
print(avg_of_normalized_loss) # mean value calculated
```

```python
# now replace missing values with new value(which is our mean).
df["normalized-losses"].replace(np.nan, avg_of_normalized_loss, inplace=True)
```

```python
df.head() # now you can see first three missing value got replaced with average value.
```

CONGRATZZ.... YOU REPLACED YOUR FIRST MISSING VALUES COLUMN.

```python
# Lets repeat same process for other columns
# Bore column
avg_bore = df['bore'].astype('float').mean(axis=0)
print(avg_bore)
```

```python
df['bore'].replace(np.nan, avg_bore, inplace=True)
```

```python
# Stroke Column
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print(avg_stroke)
```

```python
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

```python
# Horsepower
avg_hp = df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_hp, inplace=True)
```

```python
# peak-rpm
avg_rpm = df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_rpm, inplace=True)
```

```python
# num-of-doors - Now it is very intresting case. As you can not take mean of door and apply. As you can see
# doors are standard number. They can take only specific values like 2, 4, 6,8. No doors can be in 2.5, 5.6 or
# any other float values. This kind of variables are called categorical values.
df['num-of-doors'].value_counts() # Calculate grouped count
```

```python
# Now you can see, We have only two kinds of cars. 4 doors and 2 doors. If we take there mean then it will be 3
# but car with 3 doors... Wait are you serious?????
```

```python
# So in this kind of variable we take the mode value. Mode value means, most frequent value. So 4 is more frequent
# than 2 in our case.
# you can also find most frequent value using a function.
df['num-of-doors'].value_counts().idxmax()
```

```python
# So Now we have out answer. Lets replace our no-of-doors missig values with 'four'
mode_doors = df['num-of-doors'].value_counts().idxmax()
print(mode_doors)
df['num-of-doors'].replace(np.nan, mode_doors, inplace=True)
```

```python
# RHere we will drop missing value rows
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
# Now again lets check missing value colums. There should be no missing values now.
df.isnull().sum()
```

```python
# As you can see we have replaced all the missing values. Yahoooo....!!! 
```

<h2> Correct Data Format </h2>

```python
'''
if you will type df.dtypes. You will see that there are many column which are not in proper format or datatype.
e.g. price is object (it should be float.) and so on.
'''
df.dtypes
```

```python
# to get data in proper format we have to rectify there datatypes using astype().

df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

<h1>Data Standardization</h1>

```python
'''
Standardization mean changing data into common format. As we know our data was collected from different sources.
So it may have different formats e.g. 
In our dataset, the fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. Assume we are developing an application in a country that accept the fuel consumption with L/100km standard

We will need to apply data transformation to transform mpg into L/100km?

The formula for unit conversion is

L/100km = 235 / mpg
'''
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]

# rename column name from "highway-mpg" to "highway-L/100km"
#df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)

# check your transformed data 
df.head()
```

<h1>Normalization</h1>

As we know normalization means converting our data to the same scale.
We can use three differen technique to do so:
<h3>1. Simple Feature Scaling: </h3>
Xnew = X/max(X)

<h3>2. Min-Max Scaling: </h3>
    Xnew = (X-min(X))/(max(X) - min(X))

<h3>3. Z-Score Scaling: </h3>
    Xnew = (X-mean(X))/(StandardDeviation(X))

```python
# replace (original value) by (original value)/(maximum value) - Simple Feature
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df['height'] = df['height']/df['height'].max()
```

<h1>Binning</h1>

```python
#Binning means transforming continous data into discrete groups.
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
# below is the example of binning. how binning looks like?

%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(df["horsepower"])

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
# You can see how our columns(horsepower) got binned in smaller discrete groups. 50-75, 75-100, 100-125.... so on
```

```python
#In python we can acheive is using np linspace functions.
# np.linspace(min-value, max-value, no_of_bins+1)
# So in our case
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
group_names = ["Low", "Medium", "High"]
```

```python
# Creating a new column horsepower-binned which will have categorical value High, Low, Medium
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
# Group count
df["horsepower-binned"].value_counts()
```

<h1> Dummy Variables</h1>

When we ave categorical features in text form. Our system will not understand it. So we create dummy variables 
and assign then 0, 1 values.
e.g.
car is either petrol or diesel.
    <h4> FUEL_TYPE </h4>
    ---------
    Diesel
    GAS
<h4>But system doesnt know anything about gas and diesel. So we create dummy variables like</h4>
    Diesel      GAS
    1            0 (diesel car as diesel column is 1)
    0            1 (gas car as gas column is 1)
    
Dummy variables also called indicator variables.

```python
# Lets create dummy vairables for fuel-type column
dummy_var_1 = pd.get_dummies(df['fuel-type'])
dummy_var_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_var_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```

```python
df.head()
```

```python
dummy_var_aspiration = pd.get_dummies(df['aspiration'])
dummy_var_aspiration.head()
df = pd.concat([df, dummy_var_aspiration], axis = 1)

```

```python
# export final data to csv
df.to_csv('processed_car_data.csv')
```
